```python
import pandas as pd
import numpy as np
```

```python
#Leer el archivo csv
mi_df = pd.read_csv('https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/titanic.csv')
mi_df
```

```python
#Cuantos NaN existen por columna
mi_df.isnull().sum()
```

```python
#Cambiar datos NaN de la columna edad por el promedio y observar cambios
mi_df['age'].fillna((mi_df['age'].mean()),inplace= True)
mi_df.isnull().sum()
```

```python
#Cambiar datos NaN de la columna embarked
mi_df['embarked'].fillna(value='S',inplace= True)
mi_df.isnull().sum()
```

```python
#Cambiamos los datos de la columna sex usando el diccionario { 'female': 1, 'male': 0 } y usamos unique para verificar resultados
mi_df['sex'] = mi_df['sex'].replace({ 'female': 1, 'male': 0 })
mi_df['sex'].unique()
```

```python
#Cambiamos los datos de la columna embarked usando el diccionario { 'S': 0, 'C': 1, 'Q': 2 }
mi_df['embarked'] = mi_df['embarked'].replace({ 'S': 0, 'C': 1, 'Q': 2 })
mi_df['embarked'].unique()
```

```python
#Creamos la columna child que obtiene datos segun la edad 
edad_rango = [0, 16, np.inf]
edad_etiqueta = [1, 0]
mi_df['child'] = pd.cut(mi_df['age'], 
                            bins =edad_rango, 
                            labels=edad_etiqueta)
mi_df[['age','child']]
```

```python
#Eliminar las columnas ticket, cabin y name
mi_df.drop(['ticket', 'cabin' , 'name'], axis = 1, inplace = True )
mi_df
```

```python
#Validar que no existan duplicados
mi_df_dup = mi_df.duplicated()
mi_df[mi_df_dup]
```

```python
#Borramos registros duplicados 
mi_df.drop_duplicates(inplace = True)
mi_df.count()
```

```python
#Ya no hay registros duplicados
mi_df[mi_df.duplicated()]
```

```python
#Listar campos unicos del campo sex
mi_df['sex'].unique()
```

```python
mi_df
```

```python
#Quitar espacios en blancos de campos no numericos
mi_df['home.dest'] = mi_df['home.dest'].str.strip()
mi_df
```

```python
#Guardar datos limpios en un archivo json orientado a registros
mi_df.to_json('ZEMM_titanic_clean.json',orient='records')
```
